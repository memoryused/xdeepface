1. VGG-Face
ลักษณะ: พัฒนาโดย Visual Geometry Group ของ University of Oxford. ใช้โครงสร้าง Convolutional Neural Network (CNN) มีความแม่นยำสูงในงานจดจำใบหน้า.
การใช้งาน: เหมาะสำหรับงานที่ต้องการความแม่นยำสูงและมีภาพที่คุณภาพดี.

source : https://www.researchgate.net/figure/The-VGG-Face-architecture-visualizing-low-to-high-level-features-captured-as-a-facial_fig2_318798243

2. Google FaceNet
ลักษณะ: พัฒนาโดย Google, ใช้ Triplet Loss ในการฝึกฝน. มีความแม่นยำสูงและสามารถจัดการกับภาพที่มีการเปลี่ยนแปลงของประกายแสงและท่าทางได้ดี.
การใช้งาน: เหมาะสำหรับสถานการณ์ที่ภาพใบหน้ามีความหลากหลายในเงื่อนไขการถ่ายทำ.

source : https://towardsdatascience.com/a-facenet-style-approach-to-facial-recognition-dc0944efe8d1

3. OpenFace
ลักษณะ: อิงตาม Torch, โมเดลนี้เป็น open-source และใช้การฝึกฝนแบบ Triplet Loss เช่นกัน. มีประสิทธิภาพดีในขนาดโมเดลที่เล็กกว่า FaceNet.
การใช้งาน: เหมาะสำหรับการใช้งานที่มีทรัพยากรจำกัด.

source : https://cmusatyalab.github.io/openface/

4. Facebook DeepFace
ลักษณะ: พัฒนาโดย Facebook, ใช้โครงสร้าง CNN และได้รับการฝึกฝนบนฐานข้อมูลขนาดใหญ่มาก. มีความแม่นยำใกล้เคียงกับความสามารถของมนุษย์ในการจดจำใบหน้า.
การใช้งาน: เหมาะสำหรับการใช้งานที่ต้องการความแม่นยำสูงมาก.

source : https://sefiks.com/2020/02/17/face-recognition-with-facebook-deepface-in-keras/

การใช้งาน: เหมาะสำหรับสถานการณ์ที่มีการจำกัดเรื่องขนาดของภาพ.

6. Dlib

การใช้งาน: เหมาะสำหรับการใช้งานที่ต้องการเร็วและมีทรัพยากรจำกัด.

7. ArcFace
ลักษณะ: พัฒนาโดย DeepInsight, ใช้ Angular Distance เป็นเมตริกในการวัดความคล้ายคลึง. มีความแม่นยำสูงในการจดจำใบหน้าแม้ในภาพที่มีคุณภาพต่ำ.
การใช้งาน: เหมาะสำหรับสถานการณ์ที่ต้องการความแม่นยำสูงและการจัดการกับภาพที่มีคุณภาพต่ำ.

source : https://medium.com/analytics-vidhya/face-recognition-and-arcface-additive-angular-margin-loss-for-deep-face-recognition-44abc56916c

```python
from deepface import DeepFace
import cv2
```

```python
backends = [
  'opencv', 
  'ssd', 
  'dlib', 
  'mtcnn', 
  'retinaface', 
  'mediapipe',
  'yolov8',
  'yunet',
  'fastmtcnn',
]
```

```python
FACE_CASCADE = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
DB_PATH = 'data_db'
```

```python
# ภาพที่ต้องการจดจำ
path_test = 'mypic/'
img_path = 'person.jpg'
# ที่อยู่ฐานข้อมูลใบหน้า


# หาการจับคู่ที่ดีที่สุดจากฐานข้อมูล
dfs = DeepFace.find(img_path =path_test+img_path, 
        db_path = DB_PATH, 
        detector_backend = backends[1],
        model_name='VGG-Face'
)
len(dfs[0]['identity'])
```

```python
def compareImg(img):
    try :
        dfs = DeepFace.find(img_path =img, 
            db_path = DB_PATH, 
            detector_backend = backends[1],
            model_name='VGG-Face'
        )
        if len(dfs[0]['identity']) ==1:
            return dfs[0]['identity'].iloc[0],True
        else :
            return '',False
    except :
        return '',False

def face_detection(frame):
    face = FACE_CASCADE.detectMultiScale(frame,1.1,4)
    
    return face
```

```python
cap = cv2.VideoCapture(0) # 0 คือกล้องหลัก (หรือเปลี่ยนตามกล้องที่คุณใช้)

while True :
    ret, frame = cap.read()
    face= face_detection(frame)
    for (x,y,w,h) in face :
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),thickness = 5)
#         face_region = frame[y:y+h, x:x+w]
    cv2.imshow('window',frame)
    response , status = compareImg(frame)
    if status :
        print(response)
        break
# ปิดกล้อง
cap.release()
cv2.destroyAllWindows()
```
